```python
import pandas as pd
import re
```

```python
# Get story text
with open('expensivelessons.txt') as f:
  sentences = f.read().splitlines()

words = []
for i in range(len(sentences)):
  for word in sentences[i].split(" "):
    words.append({
      "sentence": i,
      "text": re.sub('[\"\(\)\,\.\;]', '', word)
    })
```

```python
# append story text to eyelink file
eyelink_data = pd.read_csv('ia_files/2.txt', sep='\t')
# for now just the first page
words_sub = words[:219]
words_df = pd.DataFrame(words_sub)
eyelink_with_text = pd.concat([words_df, eyelink_data], axis=1)
eyelink_with_text.head(20)
```
